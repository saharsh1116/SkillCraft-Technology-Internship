```python
df['Acquisition_Cost'] = (
    df['Acquisition_Cost']
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
df['Acquisition_Cost'].dtype
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'].dtype
```

```python
df.info()
```

```python
for col in ['Company', 'Campaign_Type', 'Target_Audience', 'Duration',
            'Channel_Used', 'Location', 'Language', 'Customer_Segment']:
    print(f"{col}: {df[col].unique()}")
    print()
```

```python
print("Negative Acquisition Cost:", (df['Acquisition_Cost'] < 0).sum())
print("Negative Clicks:", (df['Clicks'] < 0).sum())
print("Negative Impressions:", (df['Impressions'] < 0).sum())
print("Clicks greater than Impressions:", (df['Clicks'] > df['Impressions']).sum())
```

```python
channel_performance = df.groupby('Channel_Used').agg({
    'ROI': 'mean',
    'Conversion_Rate': 'mean',
    'Acquisition_Cost': 'mean',
    'Clicks': 'mean',
    'Impressions': 'mean',
    'Engagement_Score': 'mean'
}).round(2)

channel_performance
```

```python
import matplotlib.pyplot as plt

roi_by_channel = df.groupby('Channel_Used')['ROI'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
roi_by_channel.plot(kind='bar')

plt.title('Average ROI by Marketing Channel')
plt.xlabel('Marketing Channel')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
conversion_by_channel = df.groupby('Channel_Used')['Conversion_Rate'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
conversion_by_channel.plot(kind='bar')

plt.title('Average Conversion Rate by Marketing Channel')
plt.xlabel('Marketing Channel')
plt.ylabel('Average Conversion Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
cost_by_channel = df.groupby('Channel_Used')['Acquisition_Cost'].mean().sort_values()

plt.figure(figsize=(10, 6))
cost_by_channel.plot(kind='bar')

plt.title('Average Acquisition Cost by Marketing Channel')
plt.xlabel('Marketing Channel')
plt.ylabel('Average Acquisition Cost ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['CTR'] = (df['Clicks'] / df['Impressions']) * 100

ctr_by_channel = df.groupby('Channel_Used')['CTR'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
ctr_by_channel.plot(kind='bar')

plt.title('Average Click-Through Rate (CTR) by Marketing Channel')
plt.xlabel('Marketing Channel')
plt.ylabel('Average CTR (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Estimated_Conversions'] = df['Clicks'] * df['Conversion_Rate']

funnel_data = {
    'Impressions': df['Impressions'].sum(),
    'Clicks': df['Clicks'].sum(),
    'Estimated Conversions': df['Estimated_Conversions'].sum()
}

funnel_data
```

```python
stages = list(funnel_data.keys())
values = list(funnel_data.values())

plt.figure(figsize=(10, 6))
plt.bar(stages, values)

plt.title('Marketing Campaign Funnel')
plt.xlabel('Funnel Stage')
plt.ylabel('Total Count')
plt.tight_layout()
plt.show()
```

```python
roi_by_campaign = df.groupby('Campaign_Type')['ROI'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
roi_by_campaign.plot(kind='bar')

plt.title('Average ROI by Campaign Type')
plt.xlabel('Campaign Type')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
roi_by_segment = df.groupby('Customer_Segment')['ROI'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
roi_by_segment.plot(kind='bar')

plt.title('Average ROI by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
roi_by_audience = df.groupby('Target_Audience')['ROI'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
roi_by_audience.plot(kind='bar')

plt.title('Average ROI by Target Audience')
plt.xlabel('Target Audience')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
roi_by_location = df.groupby('Location')['ROI'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
roi_by_location.plot(kind='bar')

plt.title('Average ROI by Location')
plt.xlabel('Location')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
final_channel_analysis = df.groupby('Channel_Used').agg({
    'ROI': 'mean',
    'Conversion_Rate': 'mean',
    'Acquisition_Cost': 'mean',
    'CTR': 'mean',
    'Engagement_Score': 'mean'
}).round(3)

final_channel_analysis = final_channel_analysis.sort_values(
    by='ROI',
    ascending=False
)

final_channel_analysis
```

```python
total_impressions = df['Impressions'].sum()
total_clicks = df['Clicks'].sum()
avg_ctr = df['CTR'].mean()
avg_conversion_rate = df['Conversion_Rate'].mean() * 100
avg_roi = df['ROI'].mean()
avg_acquisition_cost = df['Acquisition_Cost'].mean()

print(f"Total Impressions: {total_impressions:,}")
print(f"Total Clicks: {total_clicks:,}")
print(f"Average CTR: {avg_ctr:.2f}%")
print(f"Average Conversion Rate: {avg_conversion_rate:.2f}%")
print(f"Average ROI: {avg_roi:.2f}")
print(f"Average Acquisition Cost: ${avg_acquisition_cost:,.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure
fig = plt.figure(figsize=(16, 10))
fig.suptitle(
    'Marketing Campaign – Business Insights Report (EDA)',
    fontsize=22,
    fontweight='bold',
    y=0.97
)

# ---------------- KPI CARDS ----------------
kpi_titles = [
    'Total Impressions',
    'Total Clicks',
    'Avg. CTR',
    'Avg. Conversion Rate',
    'Avg. ROI'
]

kpi_values = [
    f'{total_impressions/1e9:.2f}B',
    f'{total_clicks/1e6:.2f}M',
    f'{avg_ctr:.2f}%',
    f'{avg_conversion_rate:.2f}%',
    f'{avg_roi:.2f}'
]

positions = [0.04, 0.23, 0.42, 0.61, 0.80]

for x, title, value in zip(positions, kpi_titles, kpi_values):
    ax = fig.add_axes([x, 0.78, 0.16, 0.10])
    ax.text(
        0.5, 0.65, value,
        ha='center',
        va='center',
        fontsize=18,
        fontweight='bold'
    )
    ax.text(
        0.5, 0.25, title,
        ha='center',
        va='center',
        fontsize=10
    )
    ax.set_xticks([])
    ax.set_yticks([])

# ---------------- ROI BY CHANNEL ----------------
ax1 = fig.add_axes([0.06, 0.43, 0.27, 0.27])

roi_plot = df.groupby(
    'Channel_Used'
)['ROI'].mean().sort_values(ascending=False)

ax1.bar(roi_plot.index, roi_plot.values)
ax1.set_title('Average ROI by Marketing Channel')
ax1.set_ylabel('Average ROI')
ax1.tick_params(axis='x', rotation=45)

# ---------------- MARKETING FUNNEL ----------------
ax2 = fig.add_axes([0.38, 0.43, 0.27, 0.27])

funnel_labels = [
    'Impressions',
    'Clicks',
    'Est. Conversions'
]

funnel_values = [
    df['Impressions'].sum(),
    df['Clicks'].sum(),
    df['Estimated_Conversions'].sum()
]

# Display in millions for readability
funnel_millions = np.array(funnel_values) / 1e6

ax2.bar(funnel_labels, funnel_millions)
ax2.set_title('Marketing Campaign Funnel')
ax2.set_ylabel('Total (Millions)')
ax2.tick_params(axis='x', rotation=20)

# ---------------- ACQUISITION COST ----------------
ax3 = fig.add_axes([0.70, 0.43, 0.27, 0.27])

cost_plot = df.groupby(
    'Channel_Used'
)['Acquisition_Cost'].mean().sort_values()

ax3.bar(cost_plot.index, cost_plot.values)
ax3.set_title('Average Acquisition Cost by Channel')
ax3.set_ylabel('Average Cost ($)')
ax3.tick_params(axis='x', rotation=45)

# ---------------- KEY INSIGHTS ----------------
insights = (
    "KEY BUSINESS INSIGHTS\n\n"
    "• Facebook records the highest average ROI (5.019).\n"
    "• Website combines strong ROI (5.014) with the highest engagement score (5.509).\n"
    "• YouTube has the highest CTR (14.12%) and lowest acquisition cost ($12,481.39).\n"
    "• Conversion rates remain nearly identical across channels at approximately 8%."
)

fig.text(
    0.06,
    0.10,
    insights,
    fontsize=11,
    verticalalignment='bottom'
)

# ---------------- BUDGET RECOMMENDATION ----------------
recommendation = (
    "BUDGET RECOMMENDATION\n\n"
    "Prioritize moderate additional budget for Facebook and Website based on ROI and "
    "overall efficiency. Consider increasing YouTube investment due to its high CTR "
    "and low acquisition cost. Avoid drastic budget shifts because performance "
    "differences between channels are relatively small."
)

fig.text(
    0.55,
    0.10,
    recommendation,
    fontsize=11,
    verticalalignment='bottom',
    wrap=True
)

# Save report
plt.savefig(
    'Marketing_Business_Insights_Report.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```
